```python
import pandas
movieinfo = pandas.read_csv("ratings.csv")
movieinfo
```

```python
movietitles = pandas.read_csv("movietitles.csv")
movietitles
```

```python
data = pandas.merge(movieinfo, movietitles, on = 'item_id')
data
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['num_of_ratings'] = data.groupby('title')['rating'].count()
ratings.head(50)
```

```python
pivot = data.pivot_table(index= 'user_Id', columns='title', values='rating')
pivot
```

```python
selected_movie = pivot['Aladdin and the King of Thieves (1996)']
selected_movie.tail(20)
```

```python
similar = pivot.corrwith(selected_movie)
similar
```

```python
similar = pivot.corrwith(selected_movie)
similar_dataframe = pandas.DataFrame(similar, columns= ['correlation'])
similar_dataframe.sort_values('correlation', ascending=False).head(20)
```

```python
similar_dataframe = similar_dataframe.join(ratings['num_of_ratings'])
similar_dataframe[similar_dataframe['num_of_ratings']>100].sort_values('correlation', ascending=False).head(20)
```
