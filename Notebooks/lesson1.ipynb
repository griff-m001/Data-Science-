```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('bank.csv')

# Calculate education counts
education_counts = data['Education'].value_counts()

# Plot pie chart
plt.pie(education_counts, labels=education_counts.index, autopct='%1.2f%%')
plt.title('Education Distribution')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('bank.csv')

# Plot count plot
sns.countplot(data=data, x='Loan_Status')
plt.title('Loan Status Distribution')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy

# Load data
data = pd.read_csv('bank.csv')

sns.barplot(x="Gender", y="LoanAmount", palette="Oranges", data=data, estimator=numpy.mean, hue="Education")
plt.title("Education and Loan Amount")
plt.xlabel("Gender")
plt.ylabel("Education")
```

```python
#From the chart, we can see that there is a positive
# correlation between education level and average loan 
# amount. This means that individuals with higher education especially males
#  tend to have higher loan amounts on average.
```
